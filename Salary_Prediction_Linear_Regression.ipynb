# üìà Salary Prediction Using Linear Regression

This project performs a simple **linear regression** to predict an employee's **salary** based on their **years of experience**.

The dataset is widely used in introductory machine learning courses and is suitable for demonstrating:

- Data loading & exploration  
- Simple linear regression  
- Visualization of regression results  
- Interpretation of model coefficients  
- R¬≤ (coefficient of determination)

---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## üìÇ Load the Salary Dataset
The dataset contains two columns:

- **YearsExperience**
- **Salary**

Let's load and preview the first few rows.

```python
data = pd.read_csv('Salary_dataset.csv')
data.head()
```

## üîç Exploratory Data Analysis (EDA)

Before modeling, let's inspect summary statistics to understand the distribution.

```python
data.describe()
```

## üéØ Select Features (X) and Target (y)

We will predict **Salary** using **YearsExperience**.

```python
exp = data[['YearsExperience']]
salary = data[['Salary']]
```

## ü§ñ Train Linear Regression Model

We now fit a linear regression model:

- **Theta1 (slope)** ‚Äî how much salary increases per year  
- **Intercept (baseline salary)** ‚Äî expected salary at 0 years  
- **R¬≤ score** ‚Äî model accuracy  

```python
regr = LinearRegression()
regr.fit(exp, salary)

print('Theta1 : ', regr.coef_[0][0])
print('Intercept: ', regr.intercept_[0])
print('R-Square: ', regr.score(exp, salary))
```

## üìä Generate Salary Predictions

We now use the model to predict salary values to draw the regression line.

```python
predicted_salary = regr.predict(exp)
```

# üìâ Salary vs. Years of Experience ‚Äî Regression Plot

The scatter plot below shows actual data points.  
The line represents the model's predicted salary values.

```python
%matplotlib inline

plt.figure(figsize=(10,6))
plt.title('Salary vs Years of Experience', fontsize=18)
plt.xlabel('Years of Experience', fontsize=14)
plt.ylabel('Salary (USD)', fontsize=14)

plt.style.use('bmh')

plt.scatter(exp, salary, s=80, alpha=0.8, label='Actual Data')
plt.plot(exp, predicted_salary, linewidth=3, label='Regression Line')

plt.legend(fontsize=12)
plt.show()
```

# üß† Conclusion

The linear regression model shows a **strong positive correlation** between experience and salary.

### Model Summary
- **Slope:** ~ \$9,449 salary increase per additional year of experience  
- **Intercept:** ~ \$24,848  
- **R¬≤ Score:** 0.9569 (Excellent fit)

This demonstrates a clear, measurable relationship that can be used for salary forecasting and HR analytics.
